```python
#Basado en el LIBRO estadistica para administracion pagina 12
#creamos una lista vacia que que representa la edad de 100 personas
import random
edades = []
#Hacemos un ciclo para generar 100 numeros aleatorias que representan la edad de las personas
for i in range(100):
  #generamos un numero aleatorio con la funcion ranint y lo guardamos en la variable
  #numero aleatorio
  numero_aleatrio = random.randint(18,65)
  #Agregamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatrio)
print(edades)
```

```python
import pandas as pd
```

```python
#Crear un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns = ['Edades'])
df.head()
```

```python
#Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edades','Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edades')

#Imprimir la tabla de distribucion de frecuencias
print("Tabla de Distribucion de Frecuencias: ")
tabla_frecuencias
```

```python
tabla_frecuencias["Frac Relativa"] = tabla_frecuencias["Frecuencia"] / 100
tabla_frecuencias
```

```python
tabla_frecuencias["Frac Porcentaje"] = tabla_frecuencias["Frecuencia"] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias["Frac Acomulada"] = tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```

```python
#Calcular el numero de intervalos utilizando la regla de Sturges
import math
n = len(edades)
K = int(1 + math.log2(n))
print("Valor de n= ", n , "\nNumero de intervalos=", K)
```
